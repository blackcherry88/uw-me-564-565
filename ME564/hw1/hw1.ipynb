```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
_lambdas = [-5, -1, 0, 0.01, 0.1]

def exp_lambda(_lambda, t):
    return np.exp(_lambda*t)
```

```python
ts = np.linspace(0, 5, 500)
```

```python
for l in _lambdas:
    xs = exp_lambda(l, ts)
    plt.plot(ts, xs, label=f'$\\lambda={l}$')
    
plt.legend()
None
```
